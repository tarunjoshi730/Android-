**Under-Sampling**

Build a Simple Dataset Contating Similar Distribution of Normal Transaction and Fraudlent Transaction.


Number of Fraudlent Transaction Values are --> 482 ( Line in 24 )

```python
# So lets we deal with a Unbalanced Data
#more then 2 lakh Transaction in a one class
#Which In present in the normal transaction where as the 2,000 data are a
# fraud transaction So Lets we use a Method use a under sampling Method
legit_sample = legit.sample(n = 492)
```

**Concatening Two Data Frames**

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
# If we take axis 1 so its come in a column but
# If we take a row so its in good way
# whereas we use a add a two sample legit and fraud
```

```python
new_dataset.head()
```

```python
# So lets we see a All Fraudlent Transaction
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

**Spiliting the Data into Features and Targets**

```python
X = new_dataset.drop(columns='Class',axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

**Spilit the Data into Training Data and Testing Data**

```python
# So lets we use a train and testing function used in this
# whereas 0.2 is represented a 20% of data we take
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size=0.2, stratify = Y,random_state=2)
```

```python
print(X.shape,X_test.shape,X_train.shape)
```

**Model Training**

**Logistic Regression**

```python
# Logistic Regression function is used in top
model = LogisticRegression()
```

```python
# Training the Logistic Regression Model with Training Data
model.fit(X_train,Y_train)
# X is used to see a train data
# Y is used for 0 and 1
# In below its a numeric data and its can't takes a lots Time.
```

**Model Evaluation**

**Accuracy Score**

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on Training Data',training_data_accuracy)
#Lets we get a 98 percent accuracy  score
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on Testing Data',test_data_accuracy)
#Lets we get a 97 percent accuracy  score
```
